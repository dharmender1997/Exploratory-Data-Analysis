# Observation 5
Profit and sales are maximum in consumer segment and minimum in Home Office segment

```python
plt.figure(figsize=(15,10))
plt.title('Segment wise Sales in each Region',fontsize= 15)
sns.barplot(x='Region',y='Sales',data=data,hue='Segment',order=data['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

# Observation 6
Segment wise sales are almost same in every region

```python
plt.figure(figsize=(15,10))
plt.title('Segment wise Sales in each Region',fontsize= 15)
sns.barplot(x='State',y='Sales',data=data,hue='Segment',order=data['State'].value_counts().index)
plt.xlabel('Region',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
A=data.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
A.plot.bar(color=['blue','orange'],figsize=(15,8))
```
